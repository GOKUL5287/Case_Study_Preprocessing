```python
cols_for_iqr = [
    'Sale Price',
    'Flat Area (in Sqft)',
    'Lot Area (in Sqft)',
    'Area of the House from Basement (in Sqft)',
    'Basement Area (in Sqft)',
    'Living Area after Renovation (in Sqft)',
    'Lot Area after Renovation (in Sqft)'
]
# IQR based filtering is being used to IDENTIFY extreme observations
# Boxplots showed substantial outliers in area and price-related variables.
# IQR method is applied to reduce the influence of extreme observations.
```

```python
rows_before = df_house.shape[0]

for col in cols_for_iqr:

    Q1 = df_house[col].quantile(0.25)
    Q3 = df_house[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_house = df_house[(df_house[col] >= lower) & (df_house[col] <= upper)]

rows_after = df_house.shape[0]

print(f"Rows before outlier removal: {rows_before}")
print(f"Rows after outlier removal: {rows_after}")
print(f"Rows removed: {rows_before - rows_after}")
```

```python
# checking if IQR works
for columns in df_house:
    sns.boxplot(df_house[columns])
    plt.title(columns)
    plt.show()
```

```python
num_cols = df_house.select_dtypes(include=['int64', 'float64']).columns   # Not categorical, for numerical values only
# Median Imputation
# Filling missing values in numerical columns using median
for col in num_cols:
    df_house[col] = df_house[col].fillna(df_house[col].median())   # Median : Because it handles missing values better than mean, and mode is used for categorical data
```

```python
# Categorical variables do not have a meaningful mean or median. Hence using mode (most frequent category)
```

```python
cat_cols = df_house.select_dtypes(include='object').columns
# Categorical columns: 1) Date House was Sold 2)Waterfront View 3)Condition of the House
# Mode Imputation
# Filling missing values in categorical columns using mode
for col in cat_cols:
    df_house[col] = df_house[col].fillna(df_house[col].mode()[0])   # Mode : Because of categorical data
```
